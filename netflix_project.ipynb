```python
# Step 1: Split 'listed_in' column by commas and stack the genres
data_netflix_genres = data_netflix_dummy['listed_in'].str.split(',', expand=True).stack()


# Step 2: Clean up the genre entries (strip spaces and make everything lowercase)
data_netflix_genres = data_netflix_genres.str.strip().str.lower()

# Step 3: Reset the index (drop=True avoids reindexing issues)
data_netflix_genres = data_netflix_genres.reset_index(level=1, drop=True)

# Step 4: Create a new DataFrame to hold the cleaned genre data
# Important: Use the same index as data_netflix_genres after stacking
data_netflix_split = data_netflix_dummy.loc[data_netflix_genres.index].copy()  # Ensure alignment
data_netflix_split['genre'] = data_netflix_genres

# Step 5: Remove duplicate rows if necessary
data_netflix_split = data_netflix_split.drop_duplicates()

# Step 6: Group by 'genre' to get accurate counts
genre_counts = data_netflix_split.groupby('genre')['show_id'].count().reset_index(name='count')

# Step 7: Check for 'Action & Adventure'
action_adventure_count = genre_counts[genre_counts['genre'] == 'action & adventure']

# Display the count for 'Action & Adventure'
print(action_adventure_count)

# For all genres
print(genre_counts)
```
